# **4: Sliding Window Framework**

## 1: Theory

What problem does this algorithm solve?

Many array problems ask us to analyze contiguous sub-segments (subarrays or substrings) that meet a certain condition—such as finding the maximum sum of $K$ consecutive elements or the longest substring without repeating characters.

A naive approach uses nested loops to look at every possible subarray, resulting in an $O(N^2)$ or $O(N^3)$ execution time. This involves massive redundancy because the nested loops repeatedly re-sum or re-examine elements that were just looked at in the previous iteration.

The Sliding Window Framework optimizes these problems to $O(N)$ time. Instead of rebuilding a sub-segment from scratch, we maintain a "window" using two pointers (a left boundary and a right boundary). As the window slides across the data structure, we incrementally add the newly entering element on the right and remove the exiting element on the left.

There are two primary variations:

1. Fixed-Size Window: The distance between left and right is strictly locked to a constant $K$. We use this when the problem specifies an exact sub-segment length.

2. Dynamic-Size Window: The window expands or contracts dynamically based on constraints. right continually expands the window to find valid states, and left contracts it to optimize or repair broken constraints.

## 1: Fixed-Size Sliding Window

### 1: Theory

What Problem Does It Solve?

Suppose we want:

> Find the maximum sum of any 3 consecutive elements.

Array: `[2, 1, 5, 1, 3, 2]`

---

Window size: `k = 3`

Brute Force

Check every subarray:

```

[1,5,1] = 7
[5,1,3] = 9
```

This repeatedly recalculates overlapping elements.

Cost: `O(n*k)`

Sliding Window avoids recalculating everything.

````python
def max_sum_subarray(arr,k):

  # Step 1: Compute the sum of the first window of size k
  window_sum = sum(arr[:k])
  # Initialize max_sum with the first window sum
  max_sum = window_sum

  # Step 2: Slide the window across the array
  for right in range(k, len(arr)):

    # Add the next element (entering the window)
    window_sum = window_sum + arr[right]
    # Subtract the element that is leaving the window
    window_sum = window_sum - arr[right - k]
    # Update max_sum if current window_sum is larger
    max_sum = max(max_sum, window_sum)

  # Step 3: Return the maximum sum found
  return max_sum

# Test
arr = [2, 1, 5, 1, 3, 2]
k = 3
print(max_sum_subarray(arr, k))
````

### 4: Python Code
```python
def smallest_subarray(target, arr):
    left = 0
    window_sum = 0
    min_length = float("inf")

    for right in range(len(arr)):
        window_sum += arr[right]

        while window_sum >= target:
            min_length = min(min_length, right - left + 1)
            window_sum -= arr[left]
            left += 1

    return min_length
```

````python
def smallest_subarray(target, arr):
  # Initialize pointers and variables
  left = 0                # Left pointer for window start
  window_sum = 0          # Current window sum
  min_length = float("inf")  # Track smallest length found

  # Expand the window with the right pointer
  for right in range(len(arr)):
    window_sum = window_sum + arr[right] # Add current element to window sum

    # Shrink the window while condition is satisfied
    while window_sum >= target:
      # Update minimum length if current window is smaller
      min_length = min(min_length, right - left + 1)
      # Remove the leftmost element and move left pointer forward
      window_sum = window_sum - arr[left]
      left = left + 1

  # If no valid subarray found, min_length stays inf
  return min_length


arr = [2,1,5,2,3,2]
target = 7
print(smallest_subarray(target, arr))
````

### 5: Line-by-Line Explanation

`window_sum = 0`: Current sum.

`for right in range(len(arr))`: Expand window.

`window_sum += arr[right]`: Add new element.

`while window_sum >= target`:

Condition met.

Try shrinking.

`min_length = min(min_length, right - left + 1)`: Update answer.

`window_sum -= arr[left]`: Remove left element.

`left += 1`: Shrink window.

Find maximum sum of a window size: `k = 2`

Problem 2

Find average of every window: `k = 3`

````python
def average_subarray(arr, k):
  result = []
  window_sum = sum(arr[:k])  # Sum of first window
  result.append(window_sum / k)

  # Slide the window
  for right in range(k, len(arr)):
    window_sum = window_sum + arr[right]        # Add new element
    window_sum = window_sum - arr[right - k]    # Remove old element
    result.append(window_sum / k)   # Store average

  return result

# Test
arr = [1, 2, 3, 4, 5]
k = 3
print(average_subarray(arr, k))  # Expected: [2.0, 3.0, 4.0]
````

Problem 3

Find maximum element in every window of size: `2`

(Using brute-force).

````python
def max_in_windows(arr, k):
  result = []

  # Loop through each possible window
  for i in range(len(arr) - k + 1):
    # Take the slice of size k
    window = arr[i:i+k]
    # Find maximum in this window
    result.append(max(window))

  return result

# Test
arr = [1, 2, 3, 4, 5]
k = 2
print(max_in_windows(arr, k))  # Expected: [2, 3, 4, 5]
````

### Medium
Problem 4

Smallest subarray with sum ≥ target.

````python
def smallest_subarray(target, arr):
  # Initialize pointers and variables
  left = 0                # Left pointer for window start
  window_sum = 0          # Current window sum
  min_length = float("inf")  # Track smallest length found

  for right in range(len(arr)):
    window_sum += arr[right]  # Expand window

    # Shrink window while sum ≥ target
    while window_sum >= target:
      min_length = min(min_length, right - left + 1)
      window_sum = window_sum - arr[left]
      left += 1

    # If no valid subarray found, return 0
  return 0 if min_length == float("inf") else min_length


# Test
arr = [2, 1, 5, 2, 3, 2]
target = 7
print(smallest_subarray(target, arr))  # Expected output: 2 (subarray [5,2])
````

Problem 5

Longest substring without repeating characters.

````python
def longest_unique_substring(s):
  seen = set()        # Track unique characters
  left = 0            # Left pointer
  max_length = 0      # Result

  for right in range(len(s)):
    # If duplicate found, shrink window
    while s[right] in seen:
      seen.remove(s[left])
      left = left + 1

    # Add current character
    seen.add(s[right])
    # Update maximum length
    max_length = max(max_length, right - left + 1)

  return max_length

# Test
s = "abcabcbb"
print(longest_unique_substring(s))  # Expected: 3 ("abc")
s = "bbbbb"
print(longest_unique_substring(s))  # Expected: 1 ("b")
s = "pwwkew"
print(longest_unique_substring(s))  # Expected: 3 ("wke")
````

Problem 6

Maximum number of vowels in a substring of size k.

````python
def max_vowels_in_substring(s, k):
  vowels = set("aeiou")
  count = 0
  max_count = 0

  # Step 1: Count vowels in the first window
  for i in range(k):
    if s[i] in vowels:
      count = count + 1
  max_count = count

  # Step 2: Slide the window
  for right in range(k, len(s)):
    # Add new character
    if s[right] in vowels:
      count = count + 1

    # Remove old character
    if s[right - k] in vowels:
      count = count - 1

    # Update maximum
    max_count = max(max_count, count)

  return max_count


# Test
s = "abciiidef"
k = 3
print(max_vowels_in_substring(s, k))  # Expected output: 3 ("iii")
s = "leetcode"
k = 3
print(max_vowels_in_substring(s, k))  # Expected output: 2 ("lee")
s = "rhythms"
k = 4
print(max_vowels_in_substring(s, k))  # Expected output: 0 (no vowels)
````

## Assessment Level
Problem 7

Explain why Sliding Window reduces complexity from $O(N²)$ to $O(N)$.

A brute-force approach often checks every possible subarray, resulting in $O(N²)$ time complexity because each starting position may require scanning many ending positions. The Sliding Window technique avoids recomputing information for overlapping subarrays by expanding and shrinking a window as it moves through the array. Instead of restarting calculations for each new subarray, it reuses work from the previous window. Since each element enters the window once and leaves it once, the total number of operations is proportional to $N$, reducing the complexity to $O(N)$.

---

Problem 8

Why can the left pointer move at most n times?

In a Sliding Window algorithm, the left pointer only moves forward and never moves backward. Every time the window becomes invalid or needs adjustment, the left pointer advances to shrink the window. Since there are only n elements in the array, the left pointer can move from index 0 to index $N−1$ at most once. Therefore, even though it may move inside a nested loop, the total number of left-pointer movements across the entire algorithm is at most n, which helps maintain an overall $O(N)$ time complexity.

---

Problem 9

When should you use Sliding Window instead of Two Pointers?

Use the Sliding Window technique when the problem involves finding or maintaining a contiguous subarray or substring that satisfies a condition, such as a maximum sum, minimum length, or a specific set of characters. The window dynamically expands and shrinks while tracking information about the current segment. In contrast, the general Two Pointers technique is more suitable when working with sorted arrays, linked lists, or problems that involve comparing elements from different positions rather than maintaining a continuous range. In short, if the problem focuses on a contiguous range, Sliding Window is usually the preferred choice.

## Common Mistakes
Mistake 1

Forgetting to remove outgoing elements.

Wrong: `window_sum += arr[right]`

Only adding.

Never subtracting.

---

Mistake 2

Using Sliding Window when elements are revisited arbitrarily.

Sliding Window works best when data is processed sequentially.

---

Mistake 3

Using dynamic windows when a fixed window is sufficient.

This adds unnecessary complexity.

---

## Comparison: Two Pointers vs Sliding Window

| Technique           | Purpose                       |
| ------------------- | ----------------------------- |
| Two Pointers        | Compare positions             |
| Sliding Window      | Maintain a continuous segment |
| Converging Pointers | Move toward center            |
| Fast/Slow           | Different traversal speeds    |
| Fixed Window        | Constant size                 |
| Dynamic Window      | Variable size                 |
