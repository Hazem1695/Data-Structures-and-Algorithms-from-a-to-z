# **1: Contiguous Memory Layouts**

## 1: Theory

What problem does this structure solve?
When a computer executes a program, it needs a way to store lists of data so they can be accessed instantly.

Before arrays, if you needed to store 5 integers, you might have to declare 5 separate variables (`a`, `b`, `c`, `d`, `e`). This doesn't scale, and the computer might scatter those variables randomly across Random Access Memory (RAM).


An array solves this by requesting a single, unbroken (contiguous) block of memory from the operating system. Because the memory is contiguous, the computer can calculate the exact physical address of any element using a simple mathematical formula, giving us $O(1)$ random access time.


### Explain with Example:

Suppose you want to store 1 million numbers.

Without arrays:
```python
a = 10
b = 20
c = 30
...
```
This becomes impossible to manage.

Arrays solve this problem by storing many values under a single structure and allowing direct access using an index.

```python
arr = [10, 20, 30, 40, 50]
```
Access:
```python
arr[0] = 10
arr[1] = 20
arr[2] = 30
```

### Why Are Arrays Fast?

Because elements are stored in contiguous memory locations.

Contiguous means:
```Plaintext
Address: 1000 -> 10
Address: 1004 -> 20
Address: 1008 -> 30
Address: 1012 -> 40
Address: 1016 -> 50
```

## 2: Visualization

Imagine RAM as a massive chest of drawers, where each drawer has a specific numeric address.

When you allocate an array of 4 integers (assuming a 32-bit/4-byte integer size) starting at memory address `1000`, it looks like this:

```Plaintext
Index:          [0]       [1]       [2]       [3]
             +---------+---------+---------+---------+
Value:       |   42    |    7    |   19    |   88    |
             +---------+---------+---------+---------+
RAM Address:  1000      1004      1008      1012
              ^
              |__ Base Address

```

The $O(1)$ Math: To find index 2, the computer doesn't walk from the beginning. It calculates:

$$\text{Address} = \text{Base Address} + (\text{Index} \times \text{Element Size})$$

$$\text{Address} = 1000 + (2 \times 4) = 1008$$

Each integer occupies 4 bytes.

```Plaintext

1. Base address known
2. Calculate address

Cost:

```

### Insertion Cost

Suppose:
```python
[10,20,30,40]
```
Insert 15 at index 1.

Before:
```Plaintext
Index

0   1   2   3

10  20  30  40
```
Inserting Value 15 at Index 1:
```Plaintext
list after insert 15
10  15  20  30  40
```
Problem: Index 1 is currently occupied by `20`. We cannot just overwrite it

Shift Right: We must move elements from the end backward to clear a spot.

Computer must shift everything.
```Plaintext
Move 40
Move 30
Move 20
Insert 15
```
Visualization:
```Plaintext
10 20 30 40
         ↓
10 20 30 _ 40

10 20 _ 30 40

10 _ 20 30 40

10 15 20 30 40
```

Cost: $O(N)$

3 shifts + 1 write = 4 operations. (Worst case scales with array size $N$).

---

### Deletion Cost

Delete: `arr[1]`

Array: `10 20 30 40 50`

Delete `20`

Shift left:
```Plaintext
30 moves left
40 moves left
50 moves left
```
After delete: `10 30 40 50`

Cost: $O(N)$

## Complexity Analysis
| Operation             | Time Complexity |
| --------------------- | --------------- |
| Access by index       | O(1)            |
| Update by index       | O(1)            |
| Search unsorted array | O(n)            |
| Insert at end*        | O(1) amortized  |
| Insert at beginning   | O(n)            |
| Insert in middle      | O(n)            |
| Delete beginning      | O(n)            |
| Delete middle         | O(n)            |
| Delete end            | O(1)            |


## Common Mistakes
### Mistake 1

Thinking insertion is O(1)

Wrong: `Insert in middle -> O(1)`

Correct: `Insert in middle -> O(n)`

Because shifting is required.

### Mistake 2

Confusing index with address
```Plaintext
Index = logical position

Address = physical memory location

They are not the same thing.
```
### Mistake 3

Assuming arrays search quickly

`arr = [8,3,1,9,7]`

Finding: 9

requires scanning.

unless the array is sorted and binary search is used.

### Problem 4

Implement your own dynamic array using Python lists internally.

Methods:
```Plaintext
append()
insert()
delete()
get()
```

````python
class DynamicArray:
  # Constructor
  def __init__(self):
    self.data = [] # Creates an empty Python list to store elements.
  # append()
  def append(self, value):
    self.data.append(value) # Adds a new element at the end.

  # insert()
  def insert(self, index, value):
    self.data.insert(index, value) # Inserts a value at a specific position.

  # delete()
  def delete(self, index):
    if 0 <= index < len(self.data):
      del self.data[index] # Deletes the element at the given index.
    else:
      print("index out of range")

  # get()
  def get(self, index):
    if 0 <= index < len(self.data):
      return self.data[index] # Returns the value at the specified index.
    else:
      return "Index out of range"

  # show()
  def show(self):
    print(self.data) # Displays all elements.
````

#### Testing


````python
arr = DynamicArray()
# append()
arr.append(10); arr.append(20); arr.append(30)
print("array after using append method:")
arr.show()

# insert()
arr.insert(1,15)
print("array after using insert method:")
arr.show()

# delete()
arr.delete(2)
print("array after using delete method:")
arr.show()

# get()
print("array after using get method:")
print(arr.get(1))
````

### Problem 5

**Explain why binary search cannot efficiently work on a linked list.**

Binary search cannot efficiently work on a linked list because it relies on constant‑time random access to the middle element, which arrays provide but linked lists do not. In a linked list, accessing the middle node requires sequential traversal from the head, making each "middle check" an $O(N)$ operation instead of $O(1)$. As a result, the overall complexity of binary search on a linked list becomes $O(N \log N)$, which is worse than simply performing a linear search in $O(N)$. Therefore, binary search is only efficient on data structures that support direct indexing, such as arrays or dynamic arrays, but not on linked lists.
