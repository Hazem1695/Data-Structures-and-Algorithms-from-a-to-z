# **2: Brute-Force & Iterative Sorting**

## 1: Theory

What problem do these algorithms solve?
Computers process information exponentially faster when data is ordered. Searching for an item in an unsorted list of size $N$ takes $O(N)$ time because you must check every element. In a sorted list, we can use binary search, reducing it to $O(\log N)$ time.

Bubble Sort, Selection Sort, and Insertion Sort are elementary, in-place sorting algorithms. They solve the sorting problem without requiring extra helper memory, using three distinct core strategies:

1. Bubble Sort: Iteratively forces the largest unsorted element to "bubble up" to its correct position at the end of the array via adjacent swaps.

2. Selection Sort: Scans the unsorted region to find the absolute minimum element and swaps it directly into its final position at the sorting frontier.

3. Insertion Sort: Divides the array into sorted and unsorted halves, picking the next unsorted item and shifting elements backward until it fits into its correct spot (like sorting a hand of playing cards).

## 1: Bubble Sort
### 1: Theory

What Problem Does Bubble Sort Solve?

Given an unsorted array: `[5, 3, 8, 1]`

We want: `[1, 3, 5, 8]`

Bubble Sort repeatedly compares adjacent elements and swaps them if they are in the wrong order.

The largest element "bubbles" toward the end after each pass.

````python
def bubble_sort(arr):
  n = len(arr)
  for i in range(n):
    for j in range(n - 1 - i):
      if arr[j] > arr[j + 1]:
        arr[j], arr[j + 1] = arr[j + 1], arr[j]
  return arr

arr = [4, 2, 7, 1]
bubble_sort(arr)
````

## 2: Selection Sort

### 1: Theory

Instead of repeatedly swapping neighbors,

Selection Sort:

1. Finds minimum element.
2. Places it at the front.

Each pass grows the sorted region.

---

### 2: Visualization & Manual Example

Array: `[4,7,2,1]`

### Python Code


````python
def selection_sort(arr):
  n = len(arr)
  for i in range(n):
    min_index = i

    for j in range(i + 1, n):
      if arr[j] < arr[min_index]:
        min_index = j
    arr[i], arr[min_index] = arr[min_index], arr[i]

  return arr

arr = [4,7,2,1]
selection_sort(arr)
````

## 3: Insertion Sort
### 1: Theory

Insertion Sort mimics how people sort playing cards.

Maintain:
`Sorted Part | Unsorted Part`

Take one element and insert it into the correct position.

---

### 2: Visualization & Manual Example

arr = `[5,3,8,1]`

Initially:
```
[5] | [3,8,1]

Insert 3:

[3,5] | [8,1]

[3,5,8] | [1]

Insert 1:

[1,3,5,8]

sorted
```

---

### 4: Python Code

````python
def insertion_sort(arr):
  n = len(arr)

  for i in range(1, n):
    key = arr[i]
    j = i - 1

    while j >= 0 and arr[j] > key:
      arr[j + 1] = arr[j]
      j -=1

    arr[j + 1] = key
  return arr

arr = [5,3,8,1]
insertion_sort(arr)
````

#Medium

1. Modify Bubble Sort to stop early if the array becomes sorted.
2. Implement Selection Sort without Python's built-in swap syntax.
3. Implement Insertion Sort in descending order.

````python
# Modify Bubble Sort to stop early if the array becomes sorted.
def bubble_sort_optimized(arr):
  n =len(arr)

  for i in range(n):
    swapped = False  # Track if any swap happens in this pass

    for j in range(n - 1 - i):
      if arr[j] > arr[j + 1]:
        arr[j], arr[j + 1] = arr[j + 1], arr[j]
        swapped = False  # A swap accurred

    # If no two elements were swapped by the inner loop, then break
    if not swapped:
      break
  return arr

arr = [1, 2, 4, 7]
bubble_sort_optimized(arr)
````

````python
def selection_sort_standard_swap(arr):
    n = len(arr)

    for i in range(n):
      min_index = i

      for j in range(i + 1, n):
        if arr[j] < arr[min_index]:
          min_index = j

      temp = arr[i]
      arr[i] = arr[min_index]
      arr[min_index] = temp

    return arr
arr = [8, 3, 5, 1]
selection_sort_standard_swap(arr)
````

````python
# Implement Insertion Sort in descending order
def insertion_sort_descending(arr):
  n = len(arr)

  for i in range(1, n):
    key = arr[i]
    j = i - 1

    while j >= 0 and arr[j] < key:
      arr[j + 1] = arr[j]
      j -=1

    arr[j + 1] = key
  return arr

arr = [5,3,8,1]
insertion_sort_descending(arr)
````

## 1. Exact Comparisons for Bubble Sort ($N = 10$)

The exact number of comparisons Bubble Sort performs on an array of size 10 is **45**.

### The Math Behind It:

Bubble Sort's inner loop decreases its comparison count by 1 with each pass because the largest elements lock into place at the end of the array. For an array of size $N$, the formula to find the total number of comparisons is the sum of the first $N-1$ integers:

$$\text{Total Comparisons} = \frac{N \times (N - 1)}{2}$$

Plugging in $N = 10$:

$$\text{Total Comparisons} = \frac{10 \times 9}{2} = 45$$

* **Pass 1:** 9 comparisons
* **Pass 2:** 8 comparisons
* **Pass 3:** 7 comparisons
* ...and so on down to 1 comparison on the final pass ($9 + 8 + 7 + 6 + 5 + 4 + 3 + 2 + 1 = 45$).

> **Note:** This holds true for standard Bubble Sort. Even with the early-stop optimization, the *worst-case* and *average-case* number of comparisons remains exactly 45.

---

## 2. Why Selection Sort Performs Fewer Swaps Than Bubble Sort

Selection Sort is highly efficient with memory writes because it minimizes the total number of swaps to a maximum of **one swap per outer loop pass**.

* **Bubble Sort's Flaw:** It is impulsive. It swaps adjacent elements *every single time* it encounters an out-of-order pair. In a worst-case scenario, Bubble Sort can perform $O(N^2)$ swaps.
* **Selection Sort's Strategy:** It is patient. The inner loop simply updates a pointer (`min_index`) in memory as it scans the array. It performs **zero swaps** while searching. Only when the entire scan is complete does it execute a single swap to put the minimum element in its final place.

Because of this, Selection Sort performs at most $N - 1$ swaps in total, making it much lighter on memory modifications than Bubble Sort.

---

## 3. Why Insertion Sort Excels on Nearly Sorted Arrays

Insertion Sort adapts beautifully to the existing order of data, achieving a near-linear time complexity ($O(N)$) when an array is already or nearly sorted.

The secret lies in the condition of its inner `while` loop:

```python
while j >= 0 and arr[j] > key:

```

### The Mechanism:

For Insertion Sort to shift elements and do work, the sorted element to the left (`arr[j]`) must be greater than the current `key`.

1. If the array is **already sorted**, the item to the left will *never* be greater than the key.
2. The `while` loop condition immediately evaluates to `False` on the very first check.
3. The algorithm instantly skips shifting and moves to the next item.

On a nearly sorted array, the `while` loop only runs a tiny handful of times for the few misplaced elements. The rest of the time, it simply steps through the array linearly, making it faster than almost any other sorting algorithm (including Quick Sort or Merge Sort) for this specific type of data.

---

## One-Sentence Summary
1. Bubble Sort: repeatedly swaps adjacent elements → many swaps.
2. Selection Sort: finds the minimum and performs one swap per pass → few swaps.
3. Insertion Sort: only shifts misplaced elements → extremely fast when the array is already or almost sorted.
