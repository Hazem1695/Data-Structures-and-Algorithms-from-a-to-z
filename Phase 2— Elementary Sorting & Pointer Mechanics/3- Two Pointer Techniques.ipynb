# **3: Two-Pointer Techniques**

## 1: Theory

What problem does this structure or algorithm solve?

When working with linear data structures like arrays or linked lists, we often need to search for pairs, reverse elements, or detect structural traits (like cycles). A naive approach typically relies on nested loops, leading to an inefficient $O(N^2)$ time complexity.

The Two-Pointer Technique optimizes these operations to $O(N)$ time by using two independent index variables (pointers) that traverse the structure simultaneously.

There are two primary flavors:

1. **Converging Pointers (Left & Right)**: Pointers start at opposite ends of an array and move toward each other. This eliminates redundant checks when searching for pairs in sorted data or reversing structures in-place.

```
Left  →       ← Right
```
Both pointers move toward the center.


2. **Fast & Slow Pointers (Tortoise & Hare)**: Both pointers start at the same position, but move at different speeds (e.g., the fast pointer moves two steps for every one step of the slow pointer). This detects structural loops (cycles) or identifies midpoints in a single pass without using extra memory.

```
Slow → (1 step)

Fast →→ (2 steps)
```
Used mainly in linked lists and cycle detection.

## 4: Python Code

#### Palindrome Check

```python
def is_palindrome(arr):

  left = 0
  right = len(arr) - 1

  while left < right:
    if arr[left] != arr[right]:
      return false
    
    left = left + 1
    right = right - 1

  return true
```


````python
def is_palindrome(arr):

  # Initialize two pointers:
  # left starts at the beginning, right starts at the end
  left = 0
  right = len(arr) - 1

  # Loop until the two pointers meet in the middle
  while left < right:
    # If elements at left and right are not equal, it's not a palindrome
    if arr[left] != arr[right]:
      return False

    # Move left pointer forward
    left = left + 1
    # Move right pointer backward
    right = right - 1

  # If loop finishes without mismatch, array is a palindrome
  return True

# test
arr_1 = [1,2,3,2,1]
arr_2 = [1,2,3,4,1]
print(is_palindrome(arr_1))
print(is_palindrome(arr_2))
````

```python
def two_sum_sorted(arr, target):

  left = 0
  right = len(arr) - 1

  while left > right:
    current = arr[left] + arr[right]
    if current == target:
      return left, right
    elif current < target:
      left = left + 1
    else:

```

````python
def two_sum_sorted(arr, target):

  # Initialize two pointers:
  # left starts at the beginning, right starts at the end
  left = 0
  right = len(arr) - 1

  # Loop until the two pointers meet in the middle
  while left < right:
    current = arr[left] + arr[right]

    # Calculate the sum of the two current elements
    if current == target:
      return left, right
    # If sum is smaller than target, move left pointer forward
    elif current < target:
      left = left + 1
    # If sum is larger than target, move right pointer backward
    else:
      right = right - 1

  # If no pair found, return -1
  return -1

# Test
arr = [1,2,3,4,6]
target = 7

arr2 = [10,20,30,40,50]
target2 = 100

print(two_sum_sorted(arr, target))
print(two_sum_sorted(arr2, target2))
````

### 4: Python Code

Node

Cycle Detection

```python
def has_cycle(head):

    slow = head
    fast = head

    while fast and fast.next:

        slow = slow.next
        fast = fast.next.next

        if slow == fast:
            return True

    return False
```


````python
# Define a Node class for linked list
class Node:

    def __init__(self, value):
        self.value = value
        self.next = None

def has_cycle(head):
  # Initialize two pointers
  slow = head
  fast = head

  # Traverse until fast pointer reaches end
  while fast and fast.next:

    # Move slow by 1 step
    slow = slow.next
    # Move fast by 2 steps
    fast = fast.next.next

    # If they meet, cycle exists
    if slow == fast:
      return True

  # If fast reaches end, no cycle
  return False

# Test
# Create linked list: 1 -> 2 -> 3 -> 4
node1 = Node(1)
node2 = Node(2)
node3 = Node(3)
node4 = Node(4)

node1.next = node2
node2.next = node3
node3.next = node4
# No cycle here
print(has_cycle(node1))  # Expected: False

# Create cycle: 1 -> 2 -> 3 -> 4 -> back to node2
node4.next = node2
print(has_cycle(node1))  # Expected: True
````

````python
def is_palindrome(arr):

  left = 0
  right = len(arr) - 1

  while left < right:
    if arr[left] != arr[right]:
      return False

    left = left + 1
    right = right - 1

  return True

arr = [1,2,3,2,1]
print(is_palindrome(arr))
````

Problem 2

Reverse: `[10,20,30,40]`

using left and right pointers

````python
def reverse_array(arr):

    left = 0
    right = len(arr) - 1

    while left < right:
        # Swap elements at left and right
        arr[left], arr[right] = arr[right], arr[left]

        left = left + 1
        right = right - 1

    # Return the reversed array
    return arr

arr = [10, 20, 30, 40]
print(reverse_array(arr))
````

Problem 4

Implement Two Sum for a sorted array using two pointers.

````python
def two_sum_sorted(arr, target):
    # Initialize two pointers:
    # left starts at the beginning, right starts at the end
    left = 0
    right = len(arr) - 1

    # Loop until the two pointers meet
    while left < right:
        # Calculate the sum of the two current elements
        current = arr[left] + arr[right]

        # If sum matches target, return their indices
        if current == target:
            return left, right
        # If sum is smaller than target, move left pointer forward
        elif current < target:
            left = left + 1
        # If sum is larger than target, move right pointer backward
        else:
            right = right - 1

    # If no pair found, return -1
    return -1

arr = [2, 7, 11, 15]
target = 9
print(two_sum_sorted(arr, target))
````

Problem 5

Move all zeros to the end: `[0,1,0,3,12]`

using two pointers.

````python
def move_zeros(arr):
    # Initialize two pointers:
    # left points to the position to place the next non-zero
    # right scans through the array
    left = 0

    # Traverse the array with right pointer
    for right in range(len(arr)):
        # If current element is non-zero, swap it with position at left
        if arr[right] != 0:
            arr[left], arr[right] = arr[right], arr[left]
            left = left + 1  # Move left forward

    # Return the modified array with zeros moved to the end
    return arr

arr = [0, 1, 0, 3, 12]
print(move_zeros(arr))
````

Problem 6

Detect a cycle in a linked list.

````python
# Define a Node class for linked list
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None


def has_cycle(head):
    # Initialize two pointers:
    # slow moves one step, fast moves two steps
    slow = head
    fast = head

    # Traverse until fast reaches the end
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next

        # If slow and fast meet, cycle exists
        if slow == fast:
            return True

    # If fast reaches end, no cycle
    return False

# Create linked list: 1 → 2 → 3 → 4
node1 = Node(1)
node2 = Node(2)
node3 = Node(3)
node4 = Node(4)

node1.next = node2
node2.next = node3
node3.next = node4

# No cycle here
print(has_cycle(node1))

# Create cycle: 4 → back to node2
node4.next = node2
print(has_cycle(node1))
````

### Assessment Level
Problem 7

Explain why Two Sum on a sorted array can be solved in O(n) instead of O(n²).

Answer:

In an unsorted array, we may need to compare each element with every other element, leading to $O(n²)$ time complexity. However, when the array is sorted, we can use the two-pointer technique. One pointer starts at the beginning and the other at the end. If the current sum is too small, we move the left pointer rightward to increase the sum. If the sum is too large, we move the right pointer leftward to decrease it. Since each pointer moves in only one direction and never revisits an element, each element is processed at most once. Therefore, the total number of pointer movements is at most n, giving a time complexity of $O(n)$.

Problem 8

Why does the fast pointer eventually meet the slow pointer inside a cycle?

Answer:

Once both pointers enter the cycle, the fast pointer moves two steps at a time while the slow pointer moves one step. This means the fast pointer gains one node on the slow pointer during every iteration. Since the cycle contains a finite number of nodes, the distance between the two pointers continuously decreases modulo the cycle length. Eventually, the fast pointer catches up to the slow pointer, causing them to meet at some node inside the cycle. This guarantee is the key idea behind Floyd’s Cycle Detection Algorithm (Tortoise and Hare).

---

Problem 9

Prove why finding the middle node using fast/slow pointers requires only one traversal.

Answer:

The fast/slow pointer method finds the middle node by moving the slow pointer one step and the fast pointer two steps during each iteration. Because the fast pointer travels twice as fast, it reaches the end of the linked list when the slow pointer has covered only half the distance. At that moment, the slow pointer is positioned exactly at the middle node (or one of the two middle nodes in an even-length list). Since both pointers move simultaneously during a single pass through the list, the algorithm requires only one traversal, resulting in $O(n)$ time complexity and $O(1)$ extra space.
