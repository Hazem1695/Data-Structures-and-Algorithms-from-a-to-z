````python
class StaticArray:

  def __init__(self, capacity):
    # Initialize array with fixed capacity
    self.capacity = capacity
    self.size = 0
    # Pre-allocate storage
    self.data = [None] * capacity

  def append(self, value):
    # Add element at the end if space is available
    if self.size == self.capacity:
      raise Exception("Array Full") # Prevent overflow
    self.data[self.size] = value
    self.size = self.size + 1

  def get(self, index):
    # Retrieve element at given index
    if index >= self.size:
      raise IndexError
    return self.data[index]

  def show(self):
    # Print only the filled portion of the array
    print(self.data[:self.size])

# Test
arr = StaticArray(5)   # Create array with capacity 5

arr.append(10)
arr.append(20)
arr.append(30)

arr.show()             # Expected: [10, 20, 30]

print(arr.get(1))      # Expected: 20

# Try filling to capacity
arr.append(40)
arr.append(50)
arr.show()             # Expected: [10, 20, 30, 40, 50]

# Uncommenting below will raise "Array Full"
arr.append(60)
````

````python
class DynamicArray:

    def __init__(self):
        # Start with capacity = 1 (smallest possible)
        self.capacity = 1
        self.size = 0
        # Internal storage initialized with None values
        self.data = [None] * self.capacity

    def resize(self):
      # Double the capacity when array is full
      new_capacity = self.capacity * 2
      new_data = [None] * new_capacity
      # Copy old elements into new storage
      for i in range(self.size):
        new_data[i] = self.data [i]
      # Replace old storage with new one
      self.data = new_data
      self.capacity = new_capacity

    def append(self, value):
      # If array is full, resize before inserting
      if self.size == self.capacity:
        self.resize()
      # Insert new element at the end
      self.data[self.size] = value
      self.size = self.size + 1

    def get(self, index):
      # Retrieve element at index safely
      if index >= self.size:
        raise IndexError
      return self.data[index]

    def show(self):
      # Print only the filled portion of the array
      print(self.data[:self.size])
# Test
arr = DynamicArray()

# Append elements beyond initial capacity to trigger resizing
arr.append(10)
arr.append(20)   # Resizes from capacity 1 → 2
arr.append(30)   # Resizes from capacity 2 → 4
arr.append(40)
arr.append(50)   # Resizes from capacity 4 → 8

arr.show()       # Expected: [10, 20, 30, 40, 50]

print(arr.get(2))  # Expected: 30
print("Capacity:", arr.capacity)  # Expected: 8 (after resizing)
````

### 5: Line-by-Line Explanation
`def __init__(self)`: Constructor
* `self.capacity = 1`: Current allocated space.
* `self.size = 0`: Actual stored elements.
* `self.data = [None] * self.capacity`: Allocate memory block.

`def resize(self):`: Resize
* `new_capacity = self.capacity * 2`: Double memory.
* `new_data = [None] * new_capacity`: Allocate larger block.
* `for i in range(self.size):` Copy existing elements.
* `self.data = new_data`: Switch to new memory.
* `self.capacity = new_capacity`: Update capacity.

`def append(self, value):`: Append
* `if self.size == self.capacity:`: Array full?
* `self.resize()`: Grow memory.
* `self.data[self.size] = value`: Store new value.
* `self.size = self.size + 1`: Increase element count.

###Step 7: Complexity Analysis
#### ***Static Array***
| Operation     | Complexity |
| ------------- | ---------- |
| Access        | $O(1)$     |
| Append        | $O(1)$     |
| Search        | $O(N)$     |
| Insert Middle | $O(N)$     |
| Delete Middle | $O(N)$     |

#### ***Dynamic Array***
| Operation     | Complexity     |
| ------------- | -------------- |
| Access        | $O(1)$         |
| Append        | $O(1)$ amortized|
| Resize        | $(N)$          |
| Search        | $O(N)$         |
| Insert Middle | $O(N)$         |
| Delete Middle | $O(N)$         |

---

### Why Append is Amortized O(1)?

Resizes are rare:
```
1 → 2 → 4 → 8 → 16 → 32
```
Most appends: No resize needed

Therefore average cost: $O(1)$

Problem 2

What happens when inserting into a full Static Array? Explain.

***Answer***:

A static array has a fixed size that is determined when the array is created. When the array becomes full, there is no additional memory available within that array to store new elements. Therefore, inserting another element directly is impossible. To accommodate more data, a new larger array must be allocated in memory, all existing elements must be copied to the new array, and then the new element can be inserted. This copying process requires traversing all existing elements, resulting in O(n) time complexity. This limitation is one of the main reasons dynamic arrays were developed, as they automatically handle resizing when their capacity is reached.

````python
class DynamicArray:

    def __init__(self):
        # Current maximum number of elements the array can hold
        self.capacity = 1
        # Current number of elements actually stored
        self.size = 0
        # Internal storage
        self.data = [None] * self.capacity

    # Double the capacity when the array becomes full.
    def resize(self):
        # Create a new larger array
        new_capacity = self.capacity * 2
        new_data = [None] * new_capacity
        # Copy existing elements
        for i in range(self.size):
            new_data[i] = self.data[i]
        # Replace old storage
        self.data = new_data
        self.capacity = new_capacity

    # Add an element to the end of the array.
    def append(self, value):
        # Resize if no space is available
        if self.size == self.capacity:
            self.resize()
        # Insert at the next available position
        self.data[self.size] = value
        # Increase size
        self.size += 1

    # Return the value at the given index.
    def get(self, index):
        # Check if index is valid
        if index < 0 or index >= self.size:
            raise IndexError("Index out of range")
        return self.data[index]

    # Insert a value at a specific index.
    # Elements to the right are shifted one position.
    def insert(self, index, value):
        # Check if index is valid
        if index < 0 or index > self.size:
            raise IndexError("Index out of range")
        # Resize if array is full
        if self.size == self.capacity:
            self.resize()
        # Shift elements right
        for i in range(self.size, index, -1):
            self.data[i] = self.data[i - 1]
        # Insert new value
        self.data[index] = value
        # Increase size
        self.size += 1

    # Remove an element at a specific index.
    # Elements to the right are shifted left.
    def delete(self, index):
        # Check if index is valid
        if index < 0 or index >= self.size:
            raise IndexError("Index out of range")
        # Shift elements left
        for i in range(index, self.size - 1):
            self.data[i] = self.data[i + 1]
        # Clear last occupied position
        self.data[self.size - 1] = None
        # Decrease size
        self.size -= 1

    # Return the index of the first occurrence of value.
    # Return -1 if not found.
    def search(self, value):
        for i in range(self.size):
            if self.data[i] == value:
                return i
        return -1

    # Display only the used portion of the array.
    def show(self):
        print(self.data[:self.size])


# =====================================================
# TEST CASES
# =====================================================

print("===== Create Array =====")
arr = DynamicArray()

print("Initial Capacity:", arr.capacity)
print("Initial Size:", arr.size)
arr.show()

# =====================================================
# APPEND TEST
# =====================================================

print("\n===== Append Test =====")

arr.append(10)
arr.append(20)
arr.append(30)
arr.append(40)
arr.append(50)

arr.show()

print("Size:", arr.size)
print("Capacity:", arr.capacity)

# Expected:
# [10, 20, 30, 40, 50]

# =====================================================
# GET TEST
# =====================================================

print("\n===== Get Test =====")

print(arr.get(0))  # 10
print(arr.get(2))  # 30
print(arr.get(4))  # 50

# =====================================================
# INSERT TEST
# =====================================================

print("\n===== Insert Test =====")

print("Before Insert:")
arr.show()

arr.insert(2, 25)

print("After Insert 25 at index 2:")
arr.show()

# Expected:
# [10, 20, 25, 30, 40, 50]

# =====================================================
# DELETE TEST
# =====================================================

print("\n===== Delete Test =====")

print("Before Delete:")
arr.show()

arr.delete(3)

print("After Deleting Index 3:")
arr.show()

# Expected:
# [10, 20, 25, 40, 50]

# =====================================================
# SEARCH TEST
# =====================================================

print("\n===== Search Test =====")

print("Index of 25:", arr.search(25))
print("Index of 50:", arr.search(50))
print("Index of 100:", arr.search(100))

# Expected:
# 2
# 4
# -1

# =====================================================
# FINAL STATE
# =====================================================

print("\n===== Final Array =====")

arr.show()

print("Final Size:", arr.size)
print("Final Capacity:", arr.capacity)
````

This implementation demonstrates the core operations of a dynamic array exactly as they work internally in languages such as C++, Java (`ArrayList`), and Python (`list`), including manual resizing, shifting during insertion, and shifting during deletion.

Problem 7

Why is doubling capacity better than increasing capacity by one each time?

***Answer***

Doubling the capacity significantly reduces the number of resizes required as the array grows. When the capacity is doubled, resizes occur infrequently (1 → 2 → 4 → 8 → 16 ...), so most append operations take **$O(1)$** time. In contrast, increasing the capacity by only one element causes a resize on almost every insertion after the array becomes full. Each resize requires allocating a new array and copying all existing elements, which becomes very expensive. As a result, doubling capacity gives an **amortized $O(1)$** append operation, while increasing by one leads to much poorer performance.

Problem 8

Suppose capacity grows:
```
1 → 2 → 3 → 4 → 5
```
What is the total complexity of inserting n elements?

***Answer***

If the capacity increases by only one each time, every new insertion after the array is full requires allocating a new array and copying all existing elements. The total work becomes:

```text
0 + 1 + 2 + 3 + 4 + ... + (n-1)
```

This sum equals:

Since the dominant term is $(N^2)$, the total complexity of inserting **n** elements becomes **$O(N²)$**. This is much worse than the **$O(N)$** total work achieved when the capacity doubles.

Problem 9

Explain why dynamic arrays require contiguous memory but linked lists do not.

***Answer***

A dynamic array stores its elements in consecutive memory locations so that any element can be accessed directly using its index. For example, the address of an element can be calculated from the starting address and the index, giving **$O(1)$** random access. Because of this design, the entire array must occupy a single contiguous block of memory. A linked list works differently: each node stores its data and a pointer to the next node. Since nodes are connected through pointers, they can be scattered anywhere in memory and do not need to be contiguous. This flexibility makes insertion and deletion easier, but accessing the $i-th$ element requires traversal, resulting in **$O(N)$** access time.
