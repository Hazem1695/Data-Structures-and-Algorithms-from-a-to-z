# **3: Stacks (LIFO - Last In, First Out)**

## 1: Theory

What problem does this structure solve?

Computers often need to execute tasks that require keeping track of history, tracking a path backwards, or managing nested execution states. For example:

* The "Undo" Functionality: Your text editor needs to recall your most recent keystrokes in exact reverse order.

* Function Call Resolution: When Function A calls Function B, and Function B calls Function C, the CPU must remember exactly where to return when Function C finishes.

A Stack is a linear data structure governed strictly by the Last In, First Out (LIFO) principle. It restricts access so that you can only add or remove elements from one designated end, called the Top. It solves the problem of tracking history efficiently by keeping the most recent state at the front of the line, eliminating the need to search through older data.

Implementation Approaches

1. Array-Based Stack: Uses a dynamic array. Elements are appended to the end of the array, which acts as the top. Highly memory-efficient due to contiguous allocation, but can trigger occasional $O(N)$ resize costs.

2. Linked-Node Stack: Uses custom node blocks. The head pointer acts as the top of the stack. New items are inserted or removed from the front of the linked list. This ensures a guaranteed, predictable $O(1)$ time for every operation, with no resizing overhead.

````python
class StackArray:
    def __init__(self):
        self.items = []   # Use a Python list to store stack elements

    def push(self, value):
        # Add element to the top of the stack
        self.items.append(value)

    def pop(self):
        # Remove and return the top element
        if self.is_empty():
            raise Exception("Stack Empty")
        return self.items.pop()

    def peek(self):
        # Look at the top element without removing it
        if self.is_empty():
            raise Exception("Stack Empty")
        return self.items[-1]

    def is_empty(self):
        # Check if stack has no elements
        return len(self.items) == 0

    def size(self):
        # Return number of elements in stack
        return len(self.items)

    def show(self):
        # Print stack contents
        print(self.items)
# Test
stack = StackArray()
stack.push(10)
stack.push(20)
stack.push(30)

stack.show()        # Output: [10, 20, 30]
print(stack.peek()) # Output: 30 (top element)

print(stack.pop())  # Output: 30 (removed)
stack.show()        # Output: [10, 20]

print(stack.size()) # Output: 2
````

* `self.items = []`: Creates internal storage

**Push**
* `self.items.append(value)`: Insert at end. The end represents the top

**Pop**
* `self.items.pop()` :Remove last element`

**Peek**
* `self.items[-1]`: Access top without removing

#### Method 2: Stack Using Linked List

```python
class Node:

    def __init__(self, data):
        self.data = data
        self.next = None

class StackLinkedList:
    def __init__(self):
        self.top = None
        self.count = 0

    def push(self, value):
        new_node = Node(value)
        new_node.next = self.top
        self.top = new_node
        self.count += 1

    def pop(self):
        if self.top is None:
            raise Exception("Stack Empty")
        removed = self.top.data
        self.top = self.top.next
        self.count -= 1
        return removed

    def peek(self):
        if self.top is None:
            raise Exception("Stack Empty")
        return self.top.data

    def is_empty(self):
        return self.top is None

    def size(self):
        return self.count
```

````python
# Node class for linked list
class Node:
    def __init__(self, data):
        self.data = data      # Store the value
        self.next = None      # Pointer to the next node


# Stack implemented using a linked list
class StackLinkedList:
    def __init__(self):
        self.top = None       # Points to the top node of the stack
        self.count = 0        # Keep track of number of elements

    # Push: add element on top
    def push(self, value):
        new_node = Node(value)     # Create new node
        new_node.next = self.top   # Link new node to current top
        self.top = new_node        # Update top to new node
        self.count += 1

    # Pop: remove and return top element
    def pop(self):
        if self.top is None:
            raise Exception("Stack Empty")
        removed = self.top.data    # Get value of top node
        self.top = self.top.next   # Move top pointer down
        self.count -= 1
        return removed

    # Peek: look at top element without removing
    def peek(self):
        if self.top is None:
            raise Exception("Stack Empty")
        return self.top.data

    # Check if stack is empty
    def is_empty(self):
        return self.top is None

    # Return size of stack
    def size(self):
        return self.count

# Test
stack = StackLinkedList()

stack.push(10)
stack.push(20)
stack.push(30)

print("Top element:", stack.peek())   # 30
print("Size:", stack.size())          # 3

print("Pop:", stack.pop())            # 30
print("Pop:", stack.pop())            # 20

print("Is empty?", stack.is_empty())  # False
print("Size:", stack.size())          # 1

print("Pop:", stack.pop())            # 10
print("Is empty?", stack.is_empty())  # True
````

**Push**
* `new_node = Node(value)`: Create node

* `new_node.next = self.top`: Point new node to current top

* `self.top = new_node`: New node becomes top

**Pop**

* `removed = self.top.data`: Store value

* `self.top = self.top.next`: Move top pointer

* `return removed`: Return removed value

````python
class StackArray:
    def __init__(self):
        self.items = []   # Use Python list to store stack elements

    # Push: add element to the top
    def push(self, value):
        self.items.append(value)

    # Pop: remove and return top element
    def pop(self):
        if self.is_empty():
            raise Exception("Stack Empty")
        return self.items.pop()

    # Peek: look at top element without removing
    def peek(self):
        if self.is_empty():
            raise Exception("Stack Empty")
        return self.items[-1]

    # Check if stack is empty
    def is_empty(self):
        return len(self.items) == 0

    # Return size of stack
    def size(self):
        return len(self.items)

    # Show: print stack contents
    def show(self):
        print(self.items)

# Test

stack = StackArray()

# Problem 1: push(10), push(20), push(30)
stack.push(10)
stack.push(20)
stack.push(30)
print("Problem 1: Stack after pushes")
stack.show()   # Expected: [10, 20, 30]

# Problem 2: pop() twice
print("Problem 2: Pop twice")
print("Pop:", stack.pop())   # Removes 30
print("Pop:", stack.pop())   # Removes 20
stack.show()                 # Expected: [10]

# Problem 3: peek()
print("Problem 3: Peek top element")
print("Peek:", stack.peek()) # Expected: 10
stack.show()                 # Stack still: [10]
````

````python
# Node class for linked list
class Node:
    def __init__(self, data):
        self.data = data      # Store the value
        self.next = None      # Pointer to the next node


# Stack implemented using a linked list
class StackLinkedList:
    def __init__(self):
        self.top = None       # Points to the top node of the stack
        self.count = 0        # Keep track of number of elements

    # Push: add element on top
    def push(self, value):
        new_node = Node(value)     # Create new node
        new_node.next = self.top   # Link new node to current top
        self.top = new_node        # Update top to new node
        self.count += 1

    # Pop: remove and return top element
    def pop(self):
        if self.top is None:
            raise Exception("Stack Empty")
        removed = self.top.data    # Get value of top node
        self.top = self.top.next   # Move top pointer down
        self.count -= 1
        return removed

    # Peek: look at top element without removing
    def peek(self):
        if self.top is None:
            raise Exception("Stack Empty")
        return self.top.data

    # Check if stack is empty
    def is_empty(self):
        return self.top is None

    # Return size of stack
    def size(self):
        return self.count

    # Show: print stack contents (top → bottom)
    def show(self):
        current = self.top
        while current:
            print(current.data, end=" → ")
            current = current.next
        print("None")


# Test
stack = StackLinkedList()

# Problem 1: push(10), push(20), push(30)
stack.push(10)
stack.push(20)
stack.push(30)
print("Problem 1: Stack after pushes")
stack.show()   # Expected: 30 → 20 → 10 → None

# Problem 2: pop() twice
print("Problem 2: Pop twice")
print("Pop:", stack.pop())   # Removes 30
print("Pop:", stack.pop())   # Removes 20
stack.show()                 # Expected: 10 → None

# Problem 3: peek()
print("Problem 3: Peek top element")
print("Peek:", stack.peek()) # Expected: 10
stack.show()                 # Stack still: 10 → None
````

Problem 4

Reverse a string using a stack.

Input: `"HELLO"`

Output: `"OLLEH"`

````python
class StackArray:
    def __init__(self):
        self.items = []

    def push(self, value):
        self.items.append(value)

    def pop(self):
        if self.is_empty():
            raise Exception("Stack Empty")
        return self.items.pop()

    def is_empty(self):
        return len(self.items) == 0

# Function to reverse a string using stack
def reverse_string(s):
    stack = StackArray()
    # Push each character onto the stack
    for ch in s:
        stack.push(ch)
    # Pop all characters to build reversed string
    reversed_str = ""
    while not stack.is_empty():
        reversed_str += stack.pop()
    return reversed_str


# Test
input_str = "HELLO"
output_str = reverse_string(input_str)

print("Input:", input_str)
print("Output:", output_str)   # Expected: OLLEH
````

Problem 5

Check whether parentheses are balanced.

Input: `"(()())"`

Output: `True`

````python
class StackArray:
    def __init__(self):
        self.items = []

    def push(self, value):
        self.items.append(value)

    def pop(self):
        if self.is_empty():
            raise Exception("Stack Empty")
        return self.items.pop()

    def is_empty(self):
        return len(self.items) == 0

def is_balanced_parentheses(s):
    stack = StackArray()

    for ch in s:
        if ch == "(":
            stack.push(ch)
        elif ch == ")":
            if stack.is_empty():
                return False
            stack.pop()
    # If stack is empty at the end → balanced
    return stack.is_empty()


# Test
expr = "(()())"
print("Input:", expr)
print("Output:", is_balanced_parentheses(expr))  # Expected: True
````

Problem 6

Implement a stack without using Python's built-in `list.pop()` & `list.append()`.

````python
class StackManual:
    def __init__(self, capacity=100):
        self.capacity = capacity          # Maximum size of stack
        self.items = [None] * capacity    # Fixed-size array
        self.top = -1                     # Index of top element (-1 means empty)

    # Push: add element on top
    def push(self, value):
        if self.top == self.capacity - 1:
            raise Exception("Stack Overflow")
        self.top += 1
        self.items[self.top] = value

    # Pop: remove and return top element
    def pop(self):
        if self.is_empty():
            raise Exception("Stack Empty")
        value = self.items[self.top]
        self.items[self.top] = None       # Clear slot (optional)
        self.top -= 1
        return value

    # Peek: look at top element without removing
    def peek(self):
        if self.is_empty():
            raise Exception("Stack Empty")
        return self.items[self.top]

    # Check if stack is empty
    def is_empty(self):
        return self.top == -1

    # Return size of stack
    def size(self):
        return self.top + 1

    # Show: print stack contents (top → bottom)
    def show(self):
        if self.is_empty():
            print("Stack is empty")
        else:
            for i in range(self.top, -1, -1):
                print(self.items[i], end=" → ")
            print("None")

# Test
stack = StackManual(capacity=10)

stack.push(10)
stack.push(20)
stack.push(30)

print("Stack after pushes:")
stack.show()   # Output: 30 → 20 → 10 → None

print("Pop:", stack.pop())   # Removes 30
print("Pop:", stack.pop())   # Removes 20
stack.show()                 # Output: 10 → None

print("Peek:", stack.peek()) # Output: 10
print("Size:", stack.size()) # Output: 1
````

## Assessment Level

#### **Why is a stack called a LIFO data structure?**

A stack is called a **LIFO (Last In, First Out)** data structure because the last element inserted into the stack is the first one to be removed. This behavior is similar to a stack of plates: when a new plate is added, it is placed on top, and when a plate is removed, the top plate is taken first. Since all insertions (**push**) and removals (**pop**) occur at the same end, the most recently added element is always the first one accessed or removed.

---

Problem 8

#### **Explain why push and pop are $O(1)$**.

The **push** operation adds an element to the top of the stack, while the **pop** operation removes the top element. Both operations involve only a constant number of steps, such as updating an index in an array-based stack or adjusting a pointer in a linked-list stack. No traversal or shifting of elements is required. Because the amount of work remains the same regardless of the number of elements in the stack, both push and pop have a time complexity of **$O(1)$**.

---

Problem 9

#### **Why is a linked-list stack naturally suited for push/pop operations?**


A linked-list stack is naturally suited for push and pop operations because both actions occur at the head of the linked list. To **push**, a new node is created and linked to the current head, then the head pointer is updated. To **pop**, the head node is removed and the head pointer is updated to the next node. These operations require only a few pointer changes and do not involve shifting elements or resizing memory. As a result, push and pop are performed efficiently in **$O(1)$** time, making linked lists an excellent underlying structure for stacks.

## Stack vs Array vs Linked List

| Structure   | Access      | Insert End     | Delete End |
| ----------- | ----------- | -------------- | ---------- |
| Array       | $O(1)$        | $O(1)$ amortized | $O(1)$       |
| Linked List | $O(N)$        | $O(1)$ (tail)    | $O(N)$       |
| Stack       | Not Allowed | $O(1)$ Push      | $O(1)$ Pop   |
