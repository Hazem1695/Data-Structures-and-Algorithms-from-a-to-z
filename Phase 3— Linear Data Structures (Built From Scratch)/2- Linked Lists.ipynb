# **2: Linked Lists**

## 1: Intro
**What problem does this structure solve?**

Even though dynamic arrays can grow, they still rely on a single, contiguous block of memory. This creates two significant problems:
1. Expensive Insertions/Deletions at the Front/Middle: Inserting or deleting an element at index 0 requires shifting all $N$ subsequent elements in memory, costing $O(N)$ time.

2. Memory Fragmentation Risks: A dynamic array requires an unbroken, contiguous block of RAM. If you need to allocate an array for 10 million integers, but your RAM is highly fragmented (lots of small free spaces, but no single continuous block of 40MB), the allocation will fail—even if you have plenty of total free memory.

A **Linked List** solves these limitations by abandoning the requirement for contiguous memory. Instead of storing elements side-by-side, it breaks data into independent structures called Nodes. Each node contains its raw data value and a Pointer (memory address reference) to the next node in line. Because nodes can be scattered anywhere in RAM, inserting or deleting a node requires only modifying pointer references ($O(1)$ time) without shifting any data.

**Singly vs. Doubly Linked Lists**

* Singly Linked Lists: Each node points only to the next node. It uses less memory per node but can only be traversed in a forward direction.

* Doubly Linked Lists: Each node contains two pointers: one to the next node and one to the previous node. This allows full bidirectional traversal and makes deletions significantly easier, though it increases the memory footprint per node.

### 5: Python Code

**Node Class**
```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
```
**Singly Linked List**
```python
class SinglyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None

    def append(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            self.tail = new_node
            return
        self.tail.next = new_node
        self.tail = new_node

    def display(self):
        current = self.head
        while current:
            print(current.data, end=" -> ")
            current = current.next
        print("None")
```

````python
# Define a Node class to represent each element in the linked list
class Node:
    def __init__(self, data):
        self.data = data # Store the value of the node
        self.next = None # Pointer to the next node (initially None)

# Define a SinglyLinkedList class to manage the linked list
class SinglyLinkedList:
  def __init__(self):
    self.head = None    # First node in the list
    self.tail = None    # Last node in the list (helps with fast append)

  # Method to add a new node at the end of the list
  def append(self, data):
    new_node = Node(data)    # Create a new node with given data
    # If the list is empty, set both head and tail to the new node
    if self.head is None:
      self.head = new_node
      self.tail = new_node
      return
    # Otherwise, link the new node to the current tail and update tail
    self.tail.next = new_node
    self.tail = new_node

  # Method to display all nodes in the list
  def display(self):
    current = self.head    # Start from the head node
    # Traverse until the end of the list
    while current:
      print(current.data, end = ' -> ')    # Print node data
      current = current.next               # Move to the next node
    print("None")

# Test
# Create a linked list
ll = SinglyLinkedList()

# Append some values
ll.append(10)
ll.append(20)
ll.append(30)

# Display the list
ll.display()   # Expected output: 10 -> 20 -> 30 -> None
````

### 4: Python Code
#### **Doubly Node**
```python
class DNode:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None
```
#### **Doubly Linked List**
```python
class DoublyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None

    def append(self, data):
        new_node = DNode(data)
        if self.head is None:
            self.head = new_node
            self.tail = new_node
            return
        self.tail.next = new_node
        new_node.prev = self.tail
        self.tail = new_node
```

````python
# Define a Node for the doubly linked list
class DNode:
  def __init__(self, data):
    self.data = data       # Store the value of the node
    self.next = None       # Pointer to the next node
    self.prev = None       # Pointer to the previous node

# Define the DoublyLinkedList class
class DoublyLinkedList:
  def __init__(self):
    self.head = None    # First node in the list
    self.tail = None    # Last node in the list

  # Method to add a new node at the end
  def append(self, data):
    new_node = DNode(data)     # Create a new node with given data
    # If the list is empty, set both head and tail to the new node
    if self.head is None:
      self.head = new_node
      self.tail = new_node
      return
    # Otherwise, link the new node to the current tail
    self.tail.next = new_node    # Current tail points forward to new node
    new_node.prev = self.tail    # New node points back to current tail
    self.tail = new_node          # Update tail to be the new node

  # Method to display the list forward
  def display_forward(self):
    current = self.head    # Start from the head node
    while current:
      print(current.data, end=" <-> ")
      current = current.next    # Move to the next node
    print("None")

  # Method to display the list backward (using prev pointers)
  def display_backward(self):
    current = self.tail    # Start from the tail node
    while current:
      print(current.data, end=" <-> ")
      current = current.prev   # Move to the prev node
    print("None")

# Test
dll = DoublyLinkedList()
dll.append(10)
dll.append(20)
dll.append(30)

dll.display_forward()   # Output: 10 <-> 20 <-> 30 <-> None
dll.display_backward()  # Output: 30 <-> 20 <-> 10 <-> None
````

### 5: Line-by-Line Explanation

* `new_node.prev = self.tail`: Backward connection

* `self.tail.next = new_node`: Forward connection.

* `self.tail = new_node`: Move tail

````python
from typing_extensions import NoExtraItems
class Node:
  def __init__(self, data):
    self.data = data
    self.next = None

class SinglyLinkedList:
  def __init__(self):
    self.head = None
    self.tail = None

  def append(self, data):
    new_node = Node(data)
    if self.head is None:
      self.head = new_node
      self.tail = new_node
      return
    self.tail.next = new_node
    self.tail = new_node

  def insert_at_beginning(self, data):
    new_node = Node(data)       # Create new node
    new_node.next = self.head   # New node points to old head
    self.head = new_node        # Update head to new node
    if self.tail is None:       # If list was empty
        self.tail = new_node

  def delete(self, key):
    current = self.head
    prev = None
    # Traverse until we find the node
    while current and current.data != key:
        prev = current
        current = current.next
    # If node not found
    if current is None:
        return
    # If deleting the head
    if prev is None:
        self.head = current.next
    else:
        prev.next = current.next
    # If deleting the tail
    if current == self.tail:
        self.tail = prev

  # Reverse: flip the list
  def reverse(self):
    prev = None
    current = self.head
    self.tail = self.head   # Old head becomes new tail

    while current:
      next_node = current.next   # Save next node
      current.next = prev        # Reverse pointer
      prev = current             # Move prev forward
      current = next_node        # Move current forward
    self.head = prev   # Update head to last node

  def display(self):
        current = self.head
        while current:
            print(current.data, end=" → ")
            current = current.next
        print("None")

# Test
# Problem 1: Create 10 → 20 → 30
print("\nCase 1\n")
sll = SinglyLinkedList()
sll.append(10)
sll.append(20)
sll.append(30)
sll.display()   # Output: 10 → 20 → 30 → None

# Problem 2: Prepend 5
print("\nCase 2\n")
sll.insert_at_beginning(5)
sll.display()   # Output: 5 → 10 → 20 → 30 → None

# Problem 3: Delete 20
print("\nCase 3\n")
sll.delete(20)
sll.display()   # Output: 5 → 10 → 30 → None

# Problem 4: Reverse the list
print("\nCase 3\n")
sll.reverse()
sll.display()   # 30 → 10 → 5 → None
````

### Assessment Level
Problem 5

**Why is accessing index 500 in a linked list $O(N)$ but $O(1)$ in an array?**

An array stores its elements in contiguous memory locations, allowing the address of any element to be calculated directly using its index. To access index 500, the computer simply computes the memory location and retrieves the value in constant time, resulting in **O(1)** complexity. In contrast, a linked list stores elements as nodes connected by pointers. Since the nodes are not indexed, accessing the element at index 500 requires starting from the head node and following 500 links one by one until the desired node is reached. Therefore, the time required grows with the position of the element, giving a complexity of **O(N)**.

---

Problem 6

**Explain why linked lists do not require contiguous memory.**

A linked list consists of nodes where each node contains data and a pointer to the next node. Because nodes are connected through pointers, they can be stored anywhere in memory rather than occupying consecutive memory locations. When a new node is created, it only needs available memory for that single node and a pointer to link it into the list. This flexibility allows linked lists to grow dynamically without requiring a large contiguous block of memory, unlike arrays which rely on adjacent memory locations for direct indexing.

---

Problem 7

**When would you choose a linked list instead of a dynamic array?**

A linked list is a good choice when the application requires frequent insertions and deletions, especially in the middle or at the beginning of a collection. In a dynamic array, such operations often require shifting many elements, resulting in **O(N)** time. A linked list can perform these operations in **O(1)** time once the target position is known because only a few pointers need to be updated. However, linked lists sacrifice fast random access, so they are preferred when modifications are more important than indexing or searching by position.
